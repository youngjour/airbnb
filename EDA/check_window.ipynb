```python

def make_window(start_month, end_month, window_size, prediction_horizon):

    windows = []
    max_start = end_month - window_size - prediction_horizon
    for window_start in range(start_month, max_start + 1):
        window = {
            'input_start': window_start,
            'input_end': window_start + window_size,
            'target_start': window_start + window_size,
            'target_end': window_start + window_size + prediction_horizon
        }
        windows.append(window)
        
    return windows
```

```python
window_size = 6
horizon = 3
train_months = 49
val_months = 6
test_months = 7
```

```python
trains = make_window(0, train_months+(horizon-1), window_size, horizon)
vals = make_window(train_months + (horizon-1) - window_size, train_months + (horizon-1) + val_months + (horizon-1), window_size, horizon)
tests = make_window(train_months + (horizon-1) + val_months + (horizon-1) - window_size, 67, window_size, horizon)
```
