```python
import pandas as pd
import numpy as np
```

```python
air_info = pd.read_csv('../Data/Preprocessed_data/Dong/near_road_info_km.csv')
road_emb = pd.read_csv('../Data/Preprocessed_data/Dong/Road_Embeddings.csv')
hf = pd.read_csv('../Data/Preprocessed_data/Dong/Human_flow.csv')
```

```python
air_info
```

```python
air_near_road = air_info.groupby(['Reporting Month','Dong_name']).agg({'num_nearby_nodes': 'sum'})
air_near_road = air_near_road.reset_index(drop=False)
air_near_road
```

```python
idx = hf[['Reporting Month', 'Dong_name']]   # 인덱스 정보만 가져오기
road_hf = idx.merge(air_near_road, on=["Reporting Month", "Dong_name"], how="left").fillna(0)
road_hf
```

```python
road_info = road_hf.merge(road_emb, on=['Dong_name'])
road_info
```
