# raw ÏûÑÎ∞∞Îî© ÏÉùÏÑ±

```python
raw_df = new_df[['Reporting Month', 'Dong_name'] + raw_cols]
print(raw_df.shape)
print(raw_df.columns)
```

```python
import json
from collections import Counter

# Îëê Í∞ÄÏßÄ Îç∞Ïù¥ÌÑ∞ ÌòïÏãùÏùÑ Î≥ÄÌôòÌïòÎäî Ìï®Ïàò
import json

def parse_amenities(amenities):
    try:
        if isinstance(amenities, str):
            if amenities.startswith("{"):  # JSON ÌòïÏãù
                amenities_dict = json.loads(amenities)  # JSON ÌååÏã±
                return [item.lower() for sublist in amenities_dict.values() for item in sublist]  # ÏÜåÎ¨∏Ïûê Î≥ÄÌôò
            elif amenities.startswith("["):  # Î¶¨Ïä§Ìä∏ Î¨∏ÏûêÏó¥ ÌòïÏãù
                return [item.lower() for item in json.loads(amenities)]  # ÏÜåÎ¨∏Ïûê Î≥ÄÌôò
        return []
    except:
        return []

# Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôò Ï†ÅÏö©
raw_df["Parsed_Amenities"] = raw_df["Amenities"].apply(parse_amenities)
```

```python
raw_df.head()
```

```python
# Flatten & ÎπàÎèÑÏàò Í≥ÑÏÇ∞
all_amenities = [amenity for sublist in raw_df["Parsed_Amenities"] for amenity in sublist]
amenity_counts = Counter(all_amenities)
print(len(amenity_counts))

# ÏÉÅÏúÑ 50Í∞ú Amenities Ï∂úÎ†•
top_amenities = [amenity for amenity, count in amenity_counts.most_common(50)]
print("Top Amenities:", top_amenities)
```

```python
# Ï†úÏô∏Ìï† Ïó¥ÏùÑ Ï†úÏô∏Ìïú ÎÇòÎ®∏ÏßÄ Ïó¥ÏùÑ one-hot Ïù∏ÏΩîÎî©
numeric_columns = raw_df.select_dtypes(include=['number']).columns
print(numeric_columns)
excluded_columns = ['Reporting Month', 'Dong_name', 'Amenities', 'Parsed_Amenities']   # AmenitiesÎäî ÎÑàÎ¨¥ ÎßéÏïÑÏÑú Îî∞Î°ú Ï≤òÎ¶¨
excluded_columns.extend(numeric_columns)
print(excluded_columns)

encoded_df = pd.get_dummies(raw_df.drop(columns=excluded_columns), drop_first=True)

# Ï†úÏô∏Ìïú Ïó¥ÏùÑ Îã§Ïãú Ï∂îÍ∞Ä
raw_embeddings = pd.concat([raw_df[excluded_columns], encoded_df], axis=1)
raw_embeddings = raw_embeddings.drop(columns=['Amenities'])

raw_embeddings
```

```python
# Amenities ÏõêÌï´ Ïù∏ÏΩîÎî© Ï†ÅÏö©
for amenity in top_amenities:
    raw_embeddings[f"Amenity_{amenity}"] = raw_embeddings["Parsed_Amenities"].apply(lambda x: 1 if amenity in x else 0)
    
raw_embeddings = raw_embeddings.drop(columns=["Parsed_Amenities"])

# Í≤∞Í≥º ÌôïÏù∏
raw_embeddings.head()
```

```python
# Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
dong_df = raw_embeddings.copy()

dong_names = list(dong_df['Dong_name'].unique())
dong_names.append('ÏÉÅÍ≥Ñ8Îèô')
print(len(dong_names))

dong_embedding = dong_df.groupby(['Reporting Month', 'Dong_name']).mean().reset_index()
count_df = dong_df.groupby(['Reporting Month', 'Dong_name']).size().reset_index(name='count')

dong_embedding = dong_embedding.merge(count_df, on=['Reporting Month', 'Dong_name'])

print(dong_embedding.shape)
dong_embedding.head()
```

```python
date_range = pd.date_range(start='2017-01-01', end='2022-07-01', freq='MS')
date_range = date_range.strftime('%Y-%m-%d')

full_index = pd.MultiIndex.from_product([date_range, dong_names], names=['Reporting Month', 'Dong_name'])
full_df = pd.DataFrame(index=full_index).reset_index()
full_df['Reporting Month'] = pd.to_datetime(full_df['Reporting Month'], errors='coerce')
print(full_df.shape)

merged_df = pd.merge(full_df, dong_embedding, on=['Reporting Month', 'Dong_name'], how='left')
merged_df.fillna(0, inplace=True)
merged_df
```

```python
# Ï†ÄÏû•
merged_df.to_csv('../Data/Preprocessed_data/Dong/AirBnB_raw_embedding.csv', index=False)   # Í∏∞Ï°¥ AirBnB_raw_data
```
